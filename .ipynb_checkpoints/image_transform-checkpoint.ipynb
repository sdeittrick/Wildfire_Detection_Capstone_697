```python
import os
import pandas as pd
import numpy as np

# import required libaries to process images
import PIL
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline 
 
#using Keras
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

### Importing and examining the images

```python
#get list of images from corsican
dir = os.getcwd()
images = os.listdir('{}/assets/wildfire_test_set'.format(dir))
image_file = [file for file in images if '.DS_Store' not in file]
fire_image_arrays = [np.asarray(Image.open('{}/assets/wildfire_test_set/{}'.format(dir,image))) for image in image_file]
fire_label_arrays = []
for array in fire_image_arrays:
    fire_label_arrays.append([1])

#get images for fire-less mountains
images = os.listdir('{}/assets/mountain_test_set'.format(dir))
image_file = [file for file in images if '.DS_Store' not in file]
mountain_image_arrays = [np.asarray(Image.open('{}/assets/mountain_test_set/{}'.format(dir,image))) for image in image_file] 
mountain_label_arrays = []
for array in mountain_image_arrays:
    mountain_label_arrays.append([0])

all_images = fire_image_arrays+mountain_image_arrays

train_fire = fire_image_arrays[:20]
train_mountain = mountain_image_arrays[:20]
test_fire = fire_image_arrays[20:]
test_mountain = mountain_image_arrays[20:]

train_fire_labels = fire_label_arrays[:20]
train_mountain_labels = mountain_label_arrays[:20]
test_fire_labels = fire_label_arrays[20:]
test_mountain_labels = mountain_label_arrays[20:]

train_images = train_fire+train_mountain
test_images = test_fire+test_mountain
train_labels = train_fire_labels+train_mountain_labels
test_labels = test_fire_labels+test_mountain_labels

# train_images, test_images = train_images / 255.0, test_images / 255.0

train_images = [image / 255.0 for image in train_images]
test_images = [image / 255.0 for image in test_images]

train_images = np.asarray(train_images).reshape(-1,1).reshape(40,1,1,1)#.reshape(10,2,2,1)
test_images = np.asarray(test_images).reshape(-1,1)
train_labels = np.asarray(train_labels).reshape(-1,1).reshape(40,1,1,1)#.reshape(10,2,2,1)
test_labels = np.asarray(test_labels).reshape(-1,1)

# train_images = tf.convert_to_tensor(train_images, dtype=tf.float32)

print(train_images)
print(train_images.shape)
print(test_images.shape)
print(train_labels.shape)
print(test_labels.shape)


# (50000, 32, 32, 3)
# (10000, 32, 32, 3)
# (50000, 1)
# (10000, 1)
```

```python
#checking out the shape for standardization
width_means = []
length_means = []
for array in all_images:
    
#     print(array.shape)

    #width to length ratio
#     print(array.shape[0]/array.shape[1])
    
    #width & length means
    width_means.append(array.shape[0])
    length_means.append(array.shape[1])

width_mean = np.mean(width_means)
length_mean = np.mean(length_means)
    
print(' ')
print('----------------')
print('width mean')
print(width_mean)
print('----------------')
print('length mean')
print(length_mean)
```

```python
model = models.Sequential()
model.add(layers.Conv2D(1, (1,1), activation='relu', input_shape=(1,1,1)))
model.add(layers.MaxPooling2D((1,1)))
model.add(layers.Conv2D(1, (1,1), activation='relu'))
model.add(layers.MaxPooling2D((1, 1)))
model.add(layers.Conv2D(1, (1,1), activation='relu'))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(1, activation='relu'))
model.add(layers.Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))
```
